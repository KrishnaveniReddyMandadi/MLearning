```python
#Q1.Answer
#import the modules and load the data
#Data description: 8 by 8 images of handwritten images between 0 and 9.
from sklearn.datasets import load_digits
digits = load_digits()
print(digits.data.shape)
import matplotlib.pyplot as plt
for i in range(0,10):
    plt.gray()
    plt.matshow(digits.images[i], "\t")
    plt.show();
```

```python
#Q.2 answer
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import cross_val_score,train_test_split,KFold
import matplotlib.pyplot as plt
from sklearn import tree
x = digits.data
y = digits.target
X, y = load_digits(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=111)
depth = []
#max_depth parameter in DecisionTreeClassifier from 1 to 10
for i in range(1,11):
    clf = DecisionTreeClassifier(max_depth=i).fit(X_train, y_train)
    # Perform 5-fold cross validation 
    scores = cross_val_score(clf, X, y, scoring='accuracy', cv=5, n_jobs=-1)
    depth.append((scores.mean()))
print(depth)
plt.plot(depth, range(1,11), color='r')
plt.show()
```
