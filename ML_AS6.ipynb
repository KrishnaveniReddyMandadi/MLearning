1.Create 2 informative features and 1000 samples and  Visualize this dataset with a scatter plot where the color

```python
import matplotlib.pyplot as plot
# import skleran.dataset to make use of make_classification
from sklearn.datasets import make_classification
#import sklerarn to make use of svm
from sklearn import svm
from sklearn.metrics import accuracy_score
X1,Y1 = make_classification(n_samples=1000,n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=2)
figure = plot.figure()
#Title for figure
figure.suptitle('Original dataset')
plot.scatter(X1[:, 0], X1[:, 1], marker="o", c=Y1)
```

 2. Fit a Linear SVM using svm.SVC() to the synthetic dataset  and scatter plot as above but with color
c is set to the predicted values.

```python
svm_svcmodel = svm.SVC(C=1.0, kernel='linear', degree=3)
svm_svcmodel.fit(X1,Y1)
yTestPrediction = svm_svcmodel.predict(X1)
figure = plot.figure()
    #Title for figure
figure.suptitle('Actual data for predicted data')
    # Index of subplot for Testdata
plot.subplot(1, 2, 2)
    # Drawing actual plot for predicted data
plot.scatter(X1[:, 0], X1[:, 1], marker="o", c=yTestPrediction)
print(accuracy_score(Y1, yTestPrediction))
```

3. different weights change  weights=[0.25,0.75]is a 25-75 , 90: 10 , 99:01

```python
X1,Y1 = make_classification(n_samples=1000,n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=2, weights=[.75, .25])
figure = plot.figure()
#Title for figure
figure.suptitle('75: 25 split dataset')
plot.scatter(X1[:, 0], X1[:, 1], marker="o", c=Y1)
svm_svcmodel = svm.SVC(C=1.0, kernel='linear', degree=3)
svm_svcmodel.fit(X1,Y1)
yTestPrediction = svm_svcmodel.predict(X1)
figure = plot.figure()
    #Title for figure
figure.suptitle('Actual data for predicted data for 75: 25')
    # Index of subplot for Testdata
plot.subplot(1, 2, 2)
    # Drawing actual plot for predicted data
plot.scatter(X1[:, 0], X1[:, 1], marker="o", c=yTestPrediction)
print(accuracy_score(Y1, yTestPrediction))
```

4. with split 99:01 and SVC class weight = balanced

```python
X1,Y1 = make_classification(n_samples=1000,n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=2, weights=[.99, .01])
figure = plot.figure()
#Title for figure
figure.suptitle('99: 01 split dataset')
plot.scatter(X1[:, 0], X1[:, 1], marker="o", c=Y1)
svm_svcmodel = svm.SVC(C=1.0, kernel='linear', degree=3, class_weight='balanced')
svm_svcmodel.fit(X1,Y1)
yTestPrediction = svm_svcmodel.predict(X1)
figure = plot.figure()
    #Title for figure
figure.suptitle('Actual data for predicted data for 99: 01')
    # Index of subplot for Testdata
plot.subplot(1, 2, 2)
    # Drawing actual plot for predicted data
plot.scatter(X1[:, 0], X1[:, 1], marker="o", c=yTestPrediction)
print(accuracy_score(Y1, yTestPrediction))
```

Yes, the performance was balanced on setting “class_weight=balanced” makes better classification on comparing classification with 99-1 split of data
